```python
import numpy as np
import pandas as pd
import requests
import json
```

## Cocktail DB

```python
cotail_url = "https://www.thecocktaildb.com/api/json/v1/1/filter.php?c=Cocktail"

r = requests.get(cotail_url)
r
```

```python
r.text
```

```python
myjson = json.loads(r.text)
myjson['drinks'][0]['strDrink']
```

```python
cocktail_df = pd.json_normalize(myjson, record_path=['drinks'])
cocktail_df
```

## Wikipedia Virginia

```python
wikiurl = "https://en.wikipedia.org/w/api.php?action=query&list=search&srsearch=Virginia&format=json&srlimit=500"
r = requests.get(wikiurl)
r
```

```python
r.text
```

```python
wikidf = pd.json_normalize(r.json(), record_path=['query', 'search'])
wikidf
```
